```python
import pandas as pd
df = pd.read_csv('../data/train_apps_rfcd_seo_mod.csv', parse_dates=True)
```

```python
cat_columns = ['sponsor', 'grant_category', 'sponsor_c', 'grant_category_c']
for cat_column in cat_columns :
    df[cat_column] = df[cat_column].astype('category')
    df[cat_column] = df[cat_column].cat.codes
```
